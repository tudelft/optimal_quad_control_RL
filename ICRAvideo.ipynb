## Animation of training procedure

```python
from quad_race_env import *
from randomization import *
from stable_baselines3 import PPO
from quadcopter_animation import animation
import os

#  ANIMATION FUNCTION
def animate_policy(model, env, **kwargs):
    env.reset()
    def run():
        actions, _ = model.predict(env.states)  
        states, rewards, dones, infos = env.step(actions)
        # print(infos)
        return env.render()
    animation.view(run, gate_pos=env.gate_pos, gate_yaw=env.gate_yaw, **kwargs)

# the general model is: "models/nn_size_comparison/run1_64_64_64/98000000.zip"
final_index = 98 # million steps
# load n stages of training
n = 3
# indices = [int(final_index*i/n)*1000_000 for i in range(1, n+1)]
indices = np.array([1,2,10,20,98])*1000_000

# load the model
models = [PPO.load(f"models/nn_size_comparison/run1_64_64_64/{index}.zip") for index in indices]

# animate the models
os.makedirs("videos/icra_video", exist_ok=True)
os.makedirs("videos/icra_video/train", exist_ok=True)

env = Quadcopter3DGates(num_envs=100, randomization=randomization_big, pause_if_collision=True)
for i, m in enumerate(models):
    if i == len(models)-1:
        animate_policy(m, env, record_steps=500, record_file=f"videos/icra_video/train/{i}.mp4")
    else:
        animate_policy(m, env, record_steps=100, record_file=f"videos/icra_video/train/{i}.mp4")
```

```python
import os

# remove the old all.mp4 file
if os.path.exists('videos/icra_video/train/all.mp4'):
    os.remove('videos/icra_video/train/all.mp4')

# Create a text file with the list of videos
with open('videos/icra_video/train/video_list.txt', 'w') as f:
    for video_file in sorted(os.listdir('videos/icra_video/train')):
        if video_file.endswith('.mp4'):
            f.write(f"file '{os.path.abspath('videos/icra_video/train/' + video_file)}'\n")

# Run the FFmpeg command
os.system("ffmpeg -f concat -safe 0 -i videos/icra_video/train/video_list.txt -c copy videos/icra_video/train/all.mp4")
```
